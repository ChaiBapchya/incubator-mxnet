## 10,000 run of CUDA rand_int64 function call

```python
import matplotlib.pyplot as plt
```

```python
N=10000
numbers=[]
counter=0
with open("somefile.txt") as myfile:
    for line in myfile:
        if(counter==N):
            break
        numbers.append(int(line.rstrip()))
        counter+=1
#numbers
```

```python
plt.plot(numbers)
plt.ylabel('rand_int64')
plt.show()
```

```python
plt.hist(numbers,bins=16)
```
