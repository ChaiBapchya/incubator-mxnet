```python
import matplotlib.pyplot as plt
```

```python
N=100000
numbers=[]
counter=0
with open("somefile.txt") as myfile:
    for line in myfile:
        if(counter==N):
            break
        numbers.append(int(line.rstrip()))
        counter+=1
#numbers
```

```python
plt.plot(numbers)
plt.ylabel('rand_int64')
plt.show()
```

```python
bin_l=[]
for i in range(32):
    bin_l.append(i*300000000000000000)
print(bin_l)    
plt.hist(numbers,bins=bin_l)
```
